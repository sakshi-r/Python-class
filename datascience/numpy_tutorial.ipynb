```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(-10,10,0.1)
print(x.shape)
```

```python
x2= x**2
```

```python
plt.plot(x, x2)
```

```python
plt.plot?
```

```python
x3=x ** 3
```

```python
plt.figure(figsize=(5,5)) #to fix the size of the graph
plt.plot(x, label='original')
plt.plot(x2, label='squared')
plt.plot(x3, label='cubed')
plt.legend() #activate the legends
plt.xlabel("Data Points")#add a label to the y axis
plt.ylabel("Values of X") #add a label to the x axis
plt.title('The visualizatin of X and its square and cube') #add title to the figure
plt.show() #show the figure
```

```python
cities= [ 'Lucknow', 'Delhi', 'Mumbai', 'Pune', 'Kolkata']
rainfall= [39,20,12,45,78]
```

```python
plt.plot(cities, rainfall, 'o--')
```

```python
plt.bar(cities, rainfall)
plt.show()
```

```python

plt.barh(cities, rainfall)
plt.show()
```

```python
plt.figure(figsize= (10,7), facecolor='white')
plt.bar(cities, rainfall, color= 'red', edgecolor='black', hatch='//', width=[.1,.3, .2, .4, .5], align='center')
plt.xlabel("Cities")#add a label to the y axis
plt.ylabel("Rainfall") #add a label to the x axis
plt.title('The visualizatin of rainfall in various cties') #add title to the figure
plt.show() #show the figure
```

```python
plt.figure(facecolor='White')
plt.pie(rainfall, labels= cities, autopct='%1.1f%%', radius=2)
plt.show()
```

```python
plt.figure(facecolor='White')
wedges= [0,.3, 0, 0, 0]
plt.pie(rainfall, labels= cities, autopct='%1.1f%%', radius=2, explode=wedges)
plt.show()
```

```python
prices=[123, 234, 345, 456, 567, 678, 789, 890, 120, 130, 140]
qty= [12,2,3,34,4,5,56,67,54,32,6]
quality= [13,45,2,42,4,5,64,84,37,49,58]
len(prices) == len(qty) ==len(quality)
```

```python
plt.figure(facecolor='White')
plt.scatter(prices, qty, color='green', marker= 'x', aplha=.5)
plt.scatter(prices, quality, color='red', marker='o', s=100*2, alpha=.5, edgecolors='black')
plt.show()
```

```python
plt.figure(facecolor='White')
x= np.random.randint(1,10,500)
y=np.random.randint(1,10,500)
plt.hist(x, bins= 10, color='green', range=(0,10), align='left')
plt.hist(y, bins= 10, color='blue', range=(0,10), align='left', alpha= .5)
_= plt.xticks(np.arange(0,11,1))
```

```python
plt.figure(facecolor='White')
_= plt.boxplot([prices, qty, quality])
plt.xticks(np.arange(1,4), ['prices', 'qty', 'quality'], rotation= 30)
plt.show()
```
